```python
## Now we will do a billing for certain products
```

```python
print('---------------------------------------------------------')
```

```python

shopping_cart = {}
keep_shopping = True
item = str(0)
print()
total = 0

while keep_shopping:
    print()
    print('Select the operation -- >>')
    select = input("1. Add item to the cart...\n2. Remove item from the cart...\n3. Get the Bill...\n")
    
    ## Adding Item to the CART
    if select == '1' or select == 1:
        movie_no = input('Enter the SL No of the movie --->>> ')
        quantity = int(input('Enter the Quantity --->>> '))
        m = inv['Name'][movie_no]
        d = inv['Director'][movie_no]
        q = inv['Quantity'][movie_no]
        if quantity <= q:
            price = inv['Price'][movie_no]
            rating = inv['Rating'][movie_no]
            genre = inv['Genre'][movie_no]
            item = str(int(item)+1)
            print(f'The Product You Have Choosen is : {m} by {d}.\nQuantity = {quantity}')
            shopping_cart[item]={}
            shopping_cart[item]['Name'] = m
            shopping_cart[item]['Movie_Id'] = movie_no
            shopping_cart[item]['Price'] = price
            shopping_cart[item]['Quantity'] = quantity
            shopping_cart[item]['Genre'] = genre
            shopping_cart[item]['Rating'] = rating
            shopping_cart[item]['Final_Price'] = quantity * price
            
            total = total + (quantity * price)
            # recalculating the inventory
            new_quantity = int(q) - int(quantity)
            inv["Quantity"][movie_no] = new_quantity
            

            
        else:
            print(f'Sorry! We only have {q} DVDs of movie {m}')
            print('------------------------------------------------------------------------------------------------------------------')
    
    
    ## Removing Item from the CART
    if select == '2' or select == 2:
        rmv_id = input('Enter the Movie_Id to be deleted --->>> ')
        rmv_quantity = input('Enter the Quantity to be removed --->>> ')
        
        for i in shopping_cart.keys():
            j = shopping_cart[i]
            if j['Movie_Id'] == rmv_id:
                if inv["Quantity"][rmv_id] == '0' or inv["Quantity"][rmv_id] == 0:
                    shopping_cart.pop(i)
                else:
                    j['Quantity'] = j['Quantity'] - int(rmv_quantity)
                # recalculating the inventory
                new_quantity = inv["Quantity"][rmv_id] + int(rmv_quantity)
                inv["Quantity"][rmv_id] = new_quantity
                break
        
        
    
    ## Displaying the CART
    if select == '3' or select == 3:
        keep_shopping = False
        shopping_cart['Total']={}
        shopping_cart['Total']['Final_Price'] = total
        dct = pd.DataFrame.from_dict(shopping_cart)
        #print(dct)
        print('------------------------------------------------------------------------------------------------------------------')
        print('------------------------------------------------------------------------------------------------------------------')
        print('Your Shopping Cart is Ready!...')
        print()
        dt = dct.T
        dt = dt.fillna('')
        print(dt)
        print('------------------------------------------------------------------------------------------------------------------')
        print('------------------------------------------------------------------------------------------------------------------')
        print('Thank You for Shopping With Us...')
        print('------------------------------------------------------------------------------------------------------------------')
        print('------------------------------------------------------------------------------------------------------------------')
        
```

```python
## Printing The Final Bill 
```

```python
import time

bill_table = pt(['Customer Name', 'Mobile Number', 'Time of Purchase', 'Total Billing Amount (in Rs)'])
total_bill = total
bill={}

customer_name = input("Enter your name --->>> ")
mobile = input("Enter your mobile number --->>> ")
print()
current_time = time.ctime()
bill["Name"] = customer_name
bill["Mobile"] = mobile
bill["Time of Purchase"] = current_time
bill["Total Billing Amount"] = str(total_bill)

bill_table.add_row([customer_name, mobile, current_time, total_bill])
print(bill_table)
print('------------------------------------------------------------------------------------------------------------------')
print('------------------------------------------------------------------------------------------------------------------')
print('Thank You for Shopping With Us...')
print('------------------------------------------------------------------------------------------------------------------')
print('------------------------------------------------------------------------------------------------------------------')
        
```
